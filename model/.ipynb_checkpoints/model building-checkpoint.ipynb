```python
import numpy as np
import pandas as pd
```

### Convert 'Ram' , 'Weight' columns into numeric type

```python
data['Ram'] = data['Ram'].str.replace('GB','').astype('int32')
data['Weight'] = data['Weight'].str.replace('kg','').astype('float32')
```

```python
data.head()
```

### Find correlation between numeric columns and label column

```python
data.corr(numeric_only=True)['Price_euros']
```

### Company Column

```python
data['Company'].value_counts()
```

### Group company names which are having value counts< 10 into other column and otherwise as it is

```python
def add_company(inpt):
    # Get the counts of each company
    counts = data['Company'].value_counts()
    
    # Check if the count of the company is less than 10, if so return 'Other', otherwise return the company name
    if counts[inpt] < 10:
        return 'Other'
    else:
        return inpt

#Apply the function to the 'Company' column
data['Company'] = data['Company'].apply(add_company)
    
```

```python
data['Company'].value_counts()['Apple']
```

### Product Column
#### having 618 unique values doesn't impact much for prediction. So we ignore this column

```python
len(data['Product'].value_counts())
```

### ScreenResolution Column

```python
data['ScreenResolution'].value_counts()
```

### Consider only two functionalities Touchscreen and IPS

```python
#create columnn 'Touchscreen' and add 0's and 1's for each entry where the items having 'Touchscreen'. Same with the IPS
data['Touchscreen'] = data['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
data['Ips'] = data['ScreenResolution'].apply(lambda x: 1 if 'IPS' in x else 0)
```

```python
data.head()
```

### Cpu Column

```python
data['Cpu'].value_counts()
```

```python
#Each and every cpu names split by words and get first three words then again join those 3 words. Apply to the cpu column and
#create a new column with cpu name in 3 by 3 words 
data['cpu name'] = data['Cpu'].apply(lambda x : " ".join(x.split()[0:3]))
```

```python
data['cpu name'].value_counts()
```

#### Categorize all cpu types into 'Intel Core i7' 'Intel Core i5' 'Intel Core i3' 'AMD'. Remain cpu types into other

```python
def set_processor(name):
    if name =='Intel Core i7' or name == 'Intel Core i5' or name=='Intel Core i3':
        return name
    else:
        if name.split()[0] == 'AMD':
            return 'AMD'
        else :
            return 'Other'

data['cpu_name'] = data['cpu name'].apply(set_processor)
    
```

```python
data['cpu_name'].value_counts()
```

### GPU Column

```python
data['Gpu'].value_counts()
```

#### Get the first two words in gpu names and group them

```python
data['gpu name']= data['Gpu'].apply(lambda x: " ".join (x.split()[0:1]))
```

```python
data['gpu name'].value_counts()
```

```python
data.shape
```

#### Ignore the entry associated with 'ARM' gpu type as it count 1

```python
data = data[data['gpu name'] != 'ARM']
```

```python
data.shape
```

```python
data.head()
```

### OpSys Column

```python
data['OpSys'].value_counts()
```

```python
def set_os(inpt):
    if inpt == 'Windows 10' or inpt == 'Windows 7' or inpt == 'Windows 10 S ':
        return 'Windows'
    elif inpt == 'macOS' or inpt == 'Mac OS X':
        return 'Mac'
    elif inpt == 'Linux':
        return inpt
    else:
        return 'Other'

data['OpSys'] = data['OpSys'].apply(set_os)
```

```python
data['OpSys'].value_counts()
```

```python
data.tail()
```

### Drop Columns

```python
data = data.drop(columns = ['laptop_ID', 'Cpu', 'cpu name', 'Gpu', 'ScreenResolution', 'Inches','Product'])
```

```python
data.head()
```

### One hot encode the dataframe using get_dummies function

```python
data = pd.get_dummies(data)
```

```python
data.head()
```

### Convert True, False values into 0's and 1's

```python
all_columns = data.columns
### separate non-categorical columns(numerical types)
non_categorical_columns = data.select_dtypes(include=['number']).columns
### separate categorical columns
categorical_columns = [col for col in all_columns if col not in non_categorical_columns]
```

```python
data[categorical_columns].head()
```

### Converts the column x into the pandas category data type. This is a special type designed to handle categorical data efficiently.
### Once the column is a category type, cat.codes generates numerical codes for each category.

```python
data[categorical_columns] = data[categorical_columns].apply(lambda x: x.astype('category').cat.codes)
```

```python
data.head()
```

```python
x = data.drop('Price_euros', axis=1)
y = data['Price_euros']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25)
```

```python
x_train.shape, x_test.shape
```

```python
def model_acc(model):
    model.fit(x_train,y_train)
    acc = model.score(x_test, y_test)
    print(str(model)+ '--->'+ str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_acc(lr)

from sklearn.linear_model import Lasso
lasso = Lasso()
model_acc(lasso)

from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
model_acc(dt)

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
model_acc(rf)
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {'n_estimators':[10,50,100],
              'criterion':['squared_error','absolute_error','poisson']}

grid_obj = GridSearchCV(estimator=rf, param_grid=parameters)

grid_fit = grid_obj.fit(x_train,y_train)

best_model= grid_fit.best_estimator_
best_model
```

```python
best_model.score(x_test, y_test)
```

```python
import pickle
with open('predictor.pickle', 'wb') as file:
    pickle.dump(best_model, file)
```
